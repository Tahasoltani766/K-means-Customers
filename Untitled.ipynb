```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('BankCustomerData.csv')
df.head()
```

```python
list_extras = ['contact','day','month','campaign','pdays','poutcome','previous','default']
df =df.drop(list_extras,axis = 1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['housing'] = label_encoder.fit_transform(df['housing'])
df['term_deposit'] = label_encoder.fit_transform(df['term_deposit'])
df['marital'] = label_encoder.fit_transform(df['marital'])
df['loan'] = label_encoder.fit_transform(df['loan'])
df['job'] = label_encoder.fit_transform(df['job'])
df['education'] = label_encoder.fit_transform(df['education'])

df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = df.values
X = np.nan_to_num(X)

clust_dataset = StandardScaler().fit_transform(X)
clust_dataset
```

```python
cluster_num = 5
k_means = KMeans(init = "k-means++",n_clusters=cluster_num , n_init =12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df['Clus km'] = labels
df.head(10)
```

```python
df.groupby('Clus km').mean()
```

```python
x = df['age']
y = df['balance']
# plt.scatter(x = df['age'],y=df['job'])
plt.scatter(x, y, c=k_means.labels_)
plt.show()
```
